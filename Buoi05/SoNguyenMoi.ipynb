# **Số Nguyên Mới**
Cho số nguyên dương $n$ có không quá 100 chữ số. Hãy xác định số nguyên lớn nhất $m$ chia hết cho 3 và khác $n$ ở đúng một chữ số.

Ví dụ, $n = 123$ thì $m$ sẽ là $723$.

Dữ liệu: Vào từ thiết bị nhập chuẩn gồm một dòng chứa số nguyên n có không quá 100 chữ số và không chứa các số 0 không có nghĩa.

Kết quả: Đưa ra thiết bị xuất chuẩn số nguyên m tìm được. <br>

|   INPUT   |   OUTPUT  |
|-----------|-----------|
|123        |723        |

#**Abstraction**
- Tìm số nguyên m lớn nhất sao cho m chia hết cho 3 và khác n đúng 1 chữ số

# **Pattern Recognition**
> Xử lý bài toán duyệt qua các chữ số của số nguyên $n$

# **Design Algorithms**
- Ta tìm đồng dư của n với 3
- Ta xét từng trường của mod (1,2,3)
- Ta chạy vòng lặp từng chữ số trong n, so sánh với mảng mod mà ta lập ra với từng trường mod của n
- Ta thay thế 1 chữ số trong n nếu đúng dừng lại (c = 0 -> c = 1)

#**Implement**


```python
n=input()
def xuli(n):
  s=int(n)
  ds=list(n)
  mod=s%3
  b={ 1:[7,4,1],2:[8,5,2],3:[9,6,3]}
  c=0
  h=1
  while h!=0:
    for i in range (len(ds)):
      ds[i] = int(ds[i])
      for j in b[3-mod]:
        if ds[i]+j<10:
          ds[i]=ds[i]+j
          h=0
          c=1
          break
      if h==0:
        break
    h=0
  if c==0:
    if mod!=0:
            ds[-1] = int(ds[-1])-mod
    else:
            ds[-1] = int(ds[-1])-3

  return ''.join([str(x) for x in ds])
print(xuli(n))
```
